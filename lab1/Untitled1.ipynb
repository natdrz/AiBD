```python
import pandas as pd

url="https://raw.githubusercontent.com/KAIR-ISZ/public_lectures/master/Analiza%20i%20Bazy%20Danych%202021/Lab%202/Datasets/billboard.csv"
data=pd.read_csv(url,encoding='latin1')

data_melted = pd.melt(data, id_vars=["year","artist.inverted","track","time","genre","date.entered","date.peaked"],ignore_index=False,var_name="week.number", value_name="rank").dropna()
data_melted = data_melted.drop('date.peaked', 1)
data_melted.sort_index()

```
